# 2-2 파이썬 프로그래밍 기초_Matplotlib
### plotting

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10,0.1)
y_sin = np.sin(x)
```

```python
plt.plot(x,y_sin)
plt.show()
```

```python
x = np.arange(0,10,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel('x')    # x측 label
plt.ylabel('y')    # y축 label
plt.title('sin & cos')    # 그래프 제목
plt.legend(['sin','cos'])    # 범례 설정
plt.show()
```
