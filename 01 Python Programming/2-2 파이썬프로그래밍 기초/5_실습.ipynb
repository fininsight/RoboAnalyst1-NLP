# 2-2 파이썬 프로그래밍 기초_실습

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Q1.데이터 읽기
df = pd.read_csv("https://storage.googleapis.com/mledu-datasets/california_housing_train.csv", sep=",")
```

```python
# Q2.데이터 확인하기
df.describe()
```

```python
# Q3. 데이터 정렬하기
df = df.sort_values(by='households')
df
```

```python
# Q4. 데이터 얻기
df = df[(df['population'] >= 3000) & (df['median_house_value'] <= 50000)]
df
```

```python
# Q5. 평균 연산
df.mean()
```

```python
# Q6. 시각화하기
df.hist(bins=10, figsize=(20,30))
plt.show()
```

```python
# Q7. 데이터 저장
df.to_csv('house.csv', sep=',')
```
