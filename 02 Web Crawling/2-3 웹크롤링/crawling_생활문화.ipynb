```python
import requests
from bs4 import BeautifulSoup
```

```python
#네이버 생활/문화카테고리의 생활/문화 일반부분 뉴스기사
def get_url(page):
    url = 'https://news.naver.com/main/list.nhn?mode=LS2D&mid=shm&sid2=245&sid1=103&page='
    url += str(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'html.parser')
    url_ls = soup.select('div[class="list_body newsflash_body"] ul[class="type06"] dt a')
    return url_ls
```

```python
# 기사 내용 크롤링
def get_content(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'html.parser')
    for match in soup.findAll('script'):
        match.replace_with('')
        
    content_ls = soup.select('div[id="articleBodyContents"]')
        
    for content in content_ls:
        print(content.text)
        break
```

```python
# 5페이지까지 크롤링
for page in [1,2,3,4,5]:
    url_ls = get_url(page)
    for url in url_ls:
        get_content(url['href'])
```
