# 3-1 텍스트 전처리_한글

##### 전처리를 위한 기사예제 가져오기

```python
f = open("실리콘밸리기사.csv",'r',encoding='utf-8')
text = f.read()
f.close()
```

```python
print(text)
```

### 1. 토큰화

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
tokens = okt.morphs(text)

print(tokens)
```

### 2. 품사 부착

```python
oktTag = []
for token in tokens:
    oktTag += okt.pos(token)

print(oktTag)
```

### 5. 불용어 처리

```python
stopPos = ['Suffix','Punctuation','Josa','Foreign','Alpha','Number']
```

```python
# 최빈어 조회
from collections import Counter
Counter(oktTag).most_common()
```

```python
stopWord = ['의','이','를','은','과','수','했다','것','있는','한다','하는','그','있다','할','이런','되기','해야','있게','여기']
```

```python
word = []
for tag in oktTag:
    if tag[1] not in stopPos:
        if tag[0] not in stopWord:
            word.append(tag[0])
# word = ','.join(word)
```

```python
print(word)
```

### 저장하기

```python
import pandas as pd
data = pd.DataFrame([text,word])
data.to_csv('실리콘벨리기사_전처리.csv')
```
